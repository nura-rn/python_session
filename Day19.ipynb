```python
#local variable
def add():
    x=2
    y=4
    addition=x+y
    print("The addition of x and y is",addition)
    return(x,y)

num1,num2=add()
print(num1)
```

```python
#Using local variable without using return
def value():
    global girl,boy
    girl=18
    boy=19
    print(girl+boy)

value()
print(girl)
print(boy)
```

```python
n_=10
def updated():
    global n_
    n_=20
    n_=n_+100
    print("n1 value inside function is:",n_)

updated()
print(n_)
```

```python
x=eval(input("enter a number"))
y=eval(input("Enter a number"))
       
def add():
    print("The addition is",x+y)

def sub():
    print("The subtraction is",x-y)

def mul():
    print("The multiplication is",x*y)

def div():
    print("The division is",x/y)

def main():
    choice=eval(input("Enter your choice (1,2,3,4)"))
    if choice==1:
        add()

    elif choice==2:
        sub()
    elif choice==3:
        mul()

    else:
        div()

main()
    
```
